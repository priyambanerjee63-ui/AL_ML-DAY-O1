```python
# STEP 7: CHECK MISSING VALUES

print("\nMissing values before cleaning:")
print(df.isnull().sum())

# isnull() checks whether each value is missing
# sum() counts total missing values column-wise
```

```python
# STEP 8: DROP UNNECESSARY COLUMNS

# These columns are not very useful for this beginner-level preprocessing project:
# Cabin        -> too many missing values
# Name         -> text column, not directly useful for this basic model
# Ticket       -> mostly unique text values
# PassengerId  -> just an ID, not a predictive feature

drop_cols =["Cabin", "Name", "Ticket", "PassengerId"]

# some datasets may not contain all columns
# So we keep only those columns that actually exist in the DataFrame

existing_drop_cols =[col for col in drop_cols if col in df.columns]

# Drop the selected unnecessary columns
df.drop(columns=existing_drop_cols, inplace=True)

print("\nColumns after dropping unnecessary features:")
print(df.columns)
```

```python
# STEP 9: HANDLE MISSING VALUES

# Fill missing values in Age with the median
# Median is used because it is better than mean when outliers may exist

if "Age" in df.columns:
  df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing values in Embarked with the mode
# Mode means the most frequent value in that column

if "Embarked" in df.columns:
  df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode([0]))

print("\nCheck Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# STEP 10: SEPARATE FEATURES (X) AND TARGET (y)

# X contains the input features used for prediction
# y contains the output/target column that we want to predict
# Here, Survived is the target column

X = df.drop(columns=["Survived"])
y = df["Survived"]

print("\nInput features (X) preview:")
display(X.head())

print("\nTarget variable (y) preview:")
display(y.head())
```

```python
# STEP 11: APPLY ONE-HOT ENCODING TO CATEGORICAL INPUT COLUMNS

# Machine learning models work better with numeric data
# Columns like Sex and Embarked contain text values
# So we convert them into numeric format using one-hot encoding

# We only encode columns if they exist in the dataset

categorical_cols = [col for col in ["Sex", "Embarked"] if col in X.columns]


# pd.get_dummies() creates separate binary columns for each category
# drop_first=True removes one column from each category group
# This helps avoid redundant information

X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

print("\nData after one-hot encoding:")
display(X.head())

print("\nColumns after one-hot encoding:")
print(X.columns)
```

```python
# STEP 12: APPLY LABEL ENCODING TO THE TARGET COLUMN

# Label encoding converts class labels into numbers
# In this dataset, Survived is already 0 and 1 in most cases,
# but we still do this for teaching purposes

label_encoder =LabelEncoder()

y_encoded = label_encoder.fit_transform(y)

print("\nEncoded target values:")
print(y_encoded[:10])
```

```python
# STEP 13: SPLIT DATA INTO TRAINING AND TESTING SETS

# train_test_split divides the data into:
# - training data: used to train the model
# - testing data: used to evaluate the model

# test_size=0.2 means 20% data goes to testing, 80% to training
# random_state=42 ensures the split stays the same every time we run the code

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size = 0.2, random_state=42
)

print("\nTrain-Test Split Shapes:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
# Step 14: apply standardization

# Standardization changes the data so that:
# - mean becomes 0
# - standard deviation becomes 1
# This is useful when features have different ranges

# Create StandardScaler object

standard_scaler = StandardScaler()

# fit_transform() is used on training data:
# - fit() learns the mean and standard deviation from training data
# - transform() applies the scaling

X_train_standardized = standard_scaler.fit_transform(X_train)

# Only transform() is used on test data
# We do NOT fit again on test data
# This avoids data leakage

X_test_standardized = standard_scaler.transform(X_test)

print("\nFirst 5 rows after standardization:")
print(X_train_standardized[:5])
```

```python
# STEP 15: APPLY NORMALIZATION

# Normalization scales the values into a fixed range, usually 0 to 1
# This is useful when we want all features to be on the same scale

# Create MinMaxScaler object
minmax_scaler= MinMaxScaler()

# Fit and transform training data
X_train_normalized = minmax_scaler.fit_transform(X_train)

# Transform test data using the same scaler
X_test_normalized = minmax_scaler.transform(X_test)

print("\nFirst 5 rows after normalization:")
print(X_train_normalized[:5])
```
